```python
# Number of outliers
Q1 = train.quantile(0.25)
Q3 = train.quantile(0.75)
IQR = Q3 - Q1
lower_range= Q1-(1.5 * IQR)
upper_range= Q3+(1.5 * IQR)
print('Number of Outliers:')
((train < (lower_range)) | (train > (upper_range))).sum()
```

```python
print('Percentage of Outliers:')
((((train < (lower_range)) | (train > (upper_range))).sum())/17898)*100
```

#### Missing Value Imputation

```python
print('There are total 12528 observations\n')
print("Total missing values for train set:")
print('Excess kurtosis of the integrated profile: ', 1735)
print('Standard deviation of the DM-SNR curve: ', 1178)
print('Skewness of the DM-SNR curve: ', 625 ,"\n")

print("Total % of missing values for train set:")
print('Excess kurtosis of the integrated profile: ', round(((1735/12528)*100),2),'%')
print('Standard deviation of the DM-SNR curve: ', round(((1178/12528)*100),2),'%')
print('Skewness of the DM-SNR curve:: ', round(((625/12528)*100),2),'%')
```

```python
train_median = train.fillna(value=train[[' Excess kurtosis of the integrated profile',' Standard deviation of the DM-SNR curve',' Skewness of the DM-SNR curve']].median())
```

```python
train_median.isnull().sum() #null values successfully imputed
```

```python
train_final = train_median.drop([' Excess kurtosis of the integrated profile', 
                                     ' Skewness of the DM-SNR curve', 
                                     ' Standard deviation of the DM-SNR curve'], axis=1)
```

```python
train_final
```

```python
X = train_final.drop('target_class',axis=1)
y = train_final[['target_class']]
```

```python
X.columns = ['meanintprofile','stdintprofile','skewintprofile','meandmsnr','exckurtdmsnr']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

# Keeping test size as 30% of dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler() 
# Scaling the data
X_trains = ss.fit_transform(X_train)
X_tests = ss.transform(X_test)
```
